## Justificación

Estimada Sra. Zaida Abdala,

La acelerada integración de la inteligencia artificial (IA) en el ámbito del diseño web está redefiniendo los procesos creativos, la productividad y las competencias requeridas por el mercado laboral. Hoy, la IA no solo potencia la creatividad y automatiza tareas rutinarias, sino que también permite a los diseñadores y desarrolladores abordar problemas complejos, optimizar flujos de trabajo y elevar la calidad de los productos digitales[2][5][6].

Este curso, completamente estructurado en torno a las mejores herramientas gratuitas y accesibles de IA, responde a la necesidad urgente de formar profesionales capaces de dominar estos recursos desde la conceptualización hasta la publicación de sitios web. Nuestra propuesta se centra en que la propia IA sea protagonista del aprendizaje, guiando y resolviendo cada desafío del proceso, desde lo más básico (como escribir un carácter o instalar Git) hasta el diseño avanzado de interfaces y la automatización de tareas con plataformas de última generación como Firebase Studio[4].

Además, manifestamos nuestro entusiasmo y compromiso por incrementar la carga horaria futura, con el objetivo de profundizar aún más en las competencias y ofrecer una formación aún más integral y actualizada.

## Programa de Estudios  
**Duración:** 1 cuatrimestre (16 semanas)  
**Frecuencia:** 2 días por semana  
**Carga horaria:** 8 horas cátedra semanales (total: 128 hs cátedra)

### Objetivo General
Formar profesionales capaces de diseñar, prototipar y publicar sitios web funcionales y atractivos, utilizando de manera intensiva y estratégica herramientas de IA gratuitas y de vanguardia, con una visión integral y práctica orientada a la resolución de problemas reales.

---

### Unidades y Contenidos

#### **Unidad 1: Introducción a la IA en el diseño web**
- Qué es la IA y su impacto en el diseño web actual.
- Ética, límites y oportunidades de la IA en el ámbito creativo[3].
- Presentación de las mejores herramientas gratuitas y su ecosistema: ChatGPT, Figma, Canva, Uizard, Firebase Studio, entre otras[1][2][4][5][6].

#### **Unidad 2: Instalación y configuración del entorno de trabajo**
- Cómo escribir un carácter: introducción a la codificación desde cero.
- Instalación y uso básico de Git.
- Creación y configuración de cuenta en GitHub.
- Primeros pasos con editores de código gratuitos y plataformas cloud.

#### **Unidad 3: IA como asistente de aprendizaje y resolución de problemas**
- Uso de ChatGPT y asistentes similares para aprender y resolver dudas técnicas en tiempo real[2][6].
- Cómo la IA puede enseñar a usar la propia IA: prompts efectivos y aprendizaje autónomo.
- Ejercicios prácticos: resolver errores comunes, buscar documentación y optimizar procesos con IA.

#### **Unidad 4: Generación de ideas, prototipos y contenido**
- Brainstorming y generación de wireframes con IA.
- Uso de Figma y Canva potenciados por IA para diseño UI/UX[1][5].
- Creación de textos y copys efectivos con ChatGPT y otras IA de copywriting[2][6].
- Generación de imágenes, iconos y recursos visuales con Midjourney, Stable Diffusion y alternativas gratuitas[2][6].

#### **Unidad 5: Diseño y prototipado avanzado con IA**
- Prototipado rápido con Uizard, Relume y herramientas similares[2][5][6].
- Automatización de variantes de diseño y validación de ideas con IA.
- Integración de animaciones e interacciones inteligentes generadas por IA[5].

#### **Unidad 6: Desarrollo y publicación de sitios web con IA**
- Introducción a Firebase Studio: creación de proyectos, uso de plantillas y despliegue cloud[4].
- Edición de código asistida por IA: generación automática de funciones y componentes[4].
- Adaptación responsive y optimización para distintos dispositivos con IA.
- Publicación de sitios web funcionales, integración con dominios y buenas prácticas de lanzamiento[5].

#### **Unidad 7: Control de versiones y trabajo colaborativo**
- Uso de Git y GitHub para control de versiones y colaboración.
- Integración de IA para revisión y mejora de código (GitHub Copilot, Gemini, etc.)[2][4].

#### **Unidad 8: Evaluación, presentación y mejora continua**
- Presentación profesional de proyectos potenciados por IA.
- Uso de IA para feedback, testing y optimización post-lanzamiento.
- Estrategias para el aprendizaje continuo y actualización en nuevas herramientas IA.

---

### Herramientas Principales (todas con versiones gratuitas o acceso libre)
- **ChatGPT** (OpenAI, Poe.com): asistencia, generación de texto, resolución de dudas[2][6].
- **Figma**: prototipado y diseño UI/UX colaborativo[1][5].
- **Canva**: diseño gráfico rápido y accesible[1].
- **Firebase Studio**: desarrollo y despliegue cloud con IA integrada[4].
- **Git y GitHub**: control de versiones y colaboración.
- **Midjourney, Stable Diffusion**: generación de imágenes por IA (acceso limitado gratuito)[2][6].
- **Uizard, Relume**: prototipado rápido con IA[2][5][6].
- **Herramientas de copywriting IA**: Jasper, Copy.ai, entre otras (versiones gratuitas disponibles)[2][6].

---

### Metodología

- **Aprendizaje activo y basado en proyectos**: cada estudiante desarrollará un sitio web real, aplicando IA en cada etapa.
- **Resolución de problemas en tiempo real**: uso intensivo de IA como tutor y asistente.
- **Evaluación continua**: entregas parciales, feedback con IA y presentaciones finales.
- **Recursos abiertos y gratuitos**: todo el curso se apoya en herramientas accesibles para garantizar la democratización del conocimiento.
